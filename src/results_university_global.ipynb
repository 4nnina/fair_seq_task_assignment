```python
import numpy as np
import math
def avg_fairness_single(lista):
    stud_score = []

    for degree in lista:
        stud_score.append(degree)
    
    return np.mean(np.array(stud_score))
    
def avg_fairness(lista1,lista2):
    return (avg_fairness_single(lista1) + avg_fairness_single(lista2))/2

    
def std_fairness_degree(lista):
    stud_score = []

    for degree in lista:
        stud_score.append(degree)
    
    return np.std(np.array(stud_score))
    
    
def std_fairness(lista1,lista2):
    avg_prof = avg_fairness_single(lista1)
    avg_degree = avg_fairness_single(lista2)

    prof_sq_mean = 0
    for prof in lista1:
        prof_sq_mean += (prof-avg_prof)**2

    degree_sq_mean = 0
    for degree in lista2:
        degree_sq_mean += (degree-avg_degree)**2

    return math.sqrt(prof_sq_mean/len(lista1) + degree_sq_mean/len(lista2))
```
