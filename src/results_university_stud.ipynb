```python
from global_sol.data_classes_global import GlobalUniversity
from global_sol.utils_global import getBestSol, perturbate_validity
from global_sol.utils_local import getBestSolLocal

import random
import numpy as np
import pandas as pd
import copy

seed = 20

random.seed(seed)
np.random.seed(seed)
```

```python

degree1 = 'Laurea in Biotecnologie [385] Corsi di laurea - UNICO'
degree2 = 'Laurea in Informatica [420] Corsi di laurea - UNICO'
degree3 = 'Laurea in Bioinformatica [419] Corsi di laurea - UNICO'
schedule = pd.read_csv('../dataset/university/lecture_timeslots.csv')
schedule = schedule[(schedule['degree']== degree1) | (schedule['degree']== degree2) | (schedule['degree']== degree3)]

prof = schedule['prof_id'].unique().tolist()
dict_degree = {degree1:[1,2,3], degree2:[1,2,3], degree3:[1,2,3]}

obj_base = GlobalUniversity(prof, dict_degree)
obj_base = perturbate_validity(obj_base)

print(obj_base.getSingleFairnessScore())
print(obj_base.fairness_score())

#print(obj_base)
```

## Local Evaluations

```python
initialTemp = 60
finalTemp = 5
alpha = 0.5
maxPerturbation = 20
```

```python
degree_heu = getBestSolLocal(obj_base, initialTemp, finalTemp, alpha, maxPerturbation, 'cohort', True) 
print('Fairness score satisfying Students')
print(degree_heu.getSingleFairnessScore())
print(degree_heu.fairness_score())
```

# Global

```python
tmp = 35
pert = 10
```

```python
global_from_Stud = getBestSol(copy.deepcopy(degree_heu), tmp, finalTemp, alpha, pert, heuristic=True)

print("Global solution starting from the students' local solutions")
print(global_from_Stud.getSingleFairnessScore())
print(global_from_Stud.fairness_score())
```
